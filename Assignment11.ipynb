**Step 1: Importing the libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
```

**Step 2: Importing dataset**

```python
df=pd.read_csv("/content/Data (1).csv")
df
```

**Step 3: Handling the missing data**

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
dataset=df.fillna(method="ffill")
dataset.isnull().sum()
```

**Step 4: Encoding categorical data**

```python
dataset.head()
```

```python
dataset['Country'].unique()
```

```python
dataset['Purchased'].unique()
```

```python
label_encoder=preprocessing.LabelEncoder()
dataset['Country']=label_encoder.fit_transform(dataset['Country'])
dataset['Country'].unique()
```

```python
dataset['Purchased']=dataset['Purchased'].map({'Yes':1,'No':0})
dataset
```

**Step 5: Creating a dummy variable**

```python
dataset=pd.get_dummies(dataset,columns=['Country'])
dataset
```

```python
column_1=dataset.pop('Purchased')
dataset.insert(0,'Purchased',column_1)
dataset
```

**Step 6: Splitting the datasets into training sets and Test sets**

```python
x=dataset.iloc[:,[1,2,3,4,5]].values
y=dataset.iloc[:,0].values
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25)
```

**Step 7: Feature Scaling**

```python
Scaler=StandardScaler()
Scaler.fit(x_train,y_train)
x_train=Scaler.transform(x_train)
x_test=Scaler.transform(x_test)
```
